```python
import pandas as pd
import os

def create_group_csv(group_letter, student_data, filename=None):
    """
    Create a CSV file for a student group in Downloads folder
    """
    
    # Set Downloads folder path
    downloads_path = r"C:\Users\carla\Downloads"
    
    if filename is None:
        filename = f'Group{group_letter.upper()}.csv'
    
    # Full file path
    full_path = os.path.join(downloads_path, filename)
    
    # Create DataFrame
    data = []
    for student_number, student_name in student_data:
        data.append({
            'group letter': group_letter.upper(),
            'student number': student_number,
            'student name': student_name
        })
    
    df = pd.DataFrame(data)
    
    # Save to CSV in Downloads folder
    df.to_csv(full_path, index=False)
    print(f"CSV file '{full_path}' created successfully!")
    
    return df

# Example usage
if __name__ == "__main__":
    # Sample data for Group C
    group_22_students = [
        (20241646, "Carla Ferreira"),
        (20241147, "Cristiana Varela"),
        (20241137, "Gon√ßalo Bento"),
        (20241598, "Sofia Salvado"),
    ]
    
    # Create the CSV file in Downloads
    create_group_csv('22', group_22_students)
```
